# Load And Explore Data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_csv('data.csv')
data_df.head()
```

```python
prices = data_df['medianHouseValue'].to_numpy()
prices.shape
```

```python
col_names = list(data_df)[:-1]
col_names
```

```python
for col_name in col_names:
    features = data_df[col_name].to_numpy()
    plot = plt.plot(features, prices, 'ro')
    plt.setp(plot, ms=1)
    plt.title(col_name + ' - price')
    plt.xlabel(col_name)
    plt.ylabel('price')
    plt.show()
```

# PCA

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
y = data_df['medianHouseValue'].to_numpy()
y.shape
```

```python
X = data_df.drop('medianHouseValue', axis=1).to_numpy()
X.shape
```

```python
train_split = 0.8

train_count = int(train_split * len(X))
X_train = X[:train_count, :]
y_train = y[:train_count]

X_val = X[train_count:, :]
y_val = y[train_count:]

print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
n_components = 8
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(X_train_scaled)
new_X = pd.DataFrame(data = principal_components, columns = list(['PC' + str(x + 1) for x in range(n_components)]))
new_X.head()
```

```python
for i in range(1, 9):
    pc = new_X['PC' + str(i)].to_numpy()
    print(np.var(pc))
```

```python
for i in range(8):
    x = X_train_scaled[i, :]
    print(np.var(x))
```
